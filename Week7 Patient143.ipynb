```python
import pandas as pd


# Load cleaned IoT data
df = pd.read_csv("cleaned_iot_data1.csv")


# Display first few rows to verify data
print(df.head())

# Convert timestamp column to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Extract the date (YYYY-MM-DD) from the timestamp
df['date'] = df['timestamp'].dt.date

# Count occurrences by date
date_counts = df['date'].value_counts().sort_index()


import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))  # Adjust figure size
sns.lineplot(x=df["timestamp"], y=df["numeric_value"], hue=df["date"], marker="o")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)


# Add title and labels
plt.title("Heart Rate Reading", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Heart Rate Level", fontsize=12)


# Show legend
plt.legend(title="Date")


# Display the plot
plt.show()

sns.set(style="darkgrid")

plt.figure(figsize=(12, 6))  # Adjust figure size
sns.lineplot(x=df["timestamp"], y=df["numeric_value1"], hue=df["date"], marker="o")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)


# Add title and labels
plt.title("Oxygen Level Reading", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Oxygen Level", fontsize=12)


# Show legend
plt.legend(title="Date")


# Display the plot
plt.show()

sns.set(style="ticks")

plt.figure(figsize=(12, 6))  # Adjust figure size
sns.lineplot(x=df["timestamp"], y=df["numeric_value2"], hue=df["date"], marker="o")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)


# Add title and labels
plt.title("Temperature Reading", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Temperature", fontsize=12)


# Show legend
plt.legend(title="Date")


# Display the plot
plt.show()
```
